```python
import pandas as pd
import numpy as np

df_train = pd.read_excel('data/train_dataset.xlsx')
df_train.head()
```

```python
df_train_raw = df_train.copy()
```

```python
df_train = df_train.drop('Purchased', axis=1)
```

```python
df_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(df_train)
y_train = df_train_raw['Purchased']
```

```python
X_train_df = pd.DataFrame(X_train)
```

```python
X_train_df
```

```python
y_train
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=11)

parameters = {
    'max_depth':[1,2,3,4,5],
    'min_samples_split':[1,2,3]
}

grid_model = GridSearchCV(model,param_grid=parameters,cv=5)

```

```python
grid_model.fit(X_train,y_train)
```

```python
df = pd.DataFrame(grid_model.cv_results_)
df
```

```python
grid_model.best_score_
```

```python
grid_model.best_params_
```

```python
best_model = grid_model.best_estimator_
```

```python
df_test = pd.read_excel(
    "data/test_dataset.xlsx"
)
```

```python
df_test.head()
```

```python
X_test = df_test.drop('Purchased', axis=1)
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
pred = best_model.predict(X_test)
pred
```

```python
y_test = df_test['Purchased']
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(pred,y_test)
acc
```
